```python
import pandas as pd
from datetime import datetime
import json
```

```python
def load_clients():
    with open("clients.json",'r') as f:
        json_string = f.read()
        clients=json.loads(json_string)
    return clients
```

```python
def get_CR_by_month(client_id:str):
    clients=load_clients()
    client=clients[client_id]
    print(client)
    
get_CR_by_month("37246")
```
